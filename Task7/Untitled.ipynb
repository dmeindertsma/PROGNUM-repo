```python
import random

# =========================
# DATA (astronomy!)
# =========================
constellation_data = [
    {"name": "Orion", "hint1": "I am known as the hunter",
     "hint2": "I have a famous belt of three stars",
     "hint3": "I am visible in winter"},

    {"name": "Ursa Major", "hint1": "I contain the Big Dipper",
     "hint2": "I am a bear",
     "hint3": "I help find the North Star"},

    {"name": "Cassiopeia", "hint1": "I look like a W shape",
     "hint2": "I am named after a queen",
     "hint3": "I am visible in the northern sky"},

    {"name": "Scorpius", "hint1": "I represent a scorpion",
     "hint2": "I have a bright red star called Antares",
     "hint3": "I am visible in summer"},

    {"name": "Leo", "hint1": "I represent a lion",
     "hint2": "I am part of the zodiac",
     "hint3": "I am visible in spring"}
]

# =========================
# OOP CLASS
# =========================
class Constellation:
    """Class representing a constellation"""

    def __init__(self, name, hint1, hint2, hint3):
        self.name = name
        self.hints = [hint1, hint2, hint3]

    def get_hints(self):
        """Return shuffled hints"""
        random.shuffle(self.hints)
        return self.hints


# =========================
# GAME CLASS
# =========================
class GuessGame:
    """Main game class"""

    def __init__(self, data):
        self.constellations = [
            Constellation(d["name"], d["hint1"], d["hint2"], d["hint3"])
            for d in data
        ]
        self.score = 0

    def play_round(self):
        """Play one round"""
        constellation = random.choice(self.constellations)
        hints = constellation.get_hints()

        print("\n🌌 Guess the Constellation!")

        for hint in hints:
            print("Hint:", hint)
            guess = input("Your guess: ")

            if guess.lower() == constellation.name.lower():
                print("✅ Correct!")
                self.score += 1
                return
            else:
                print("❌ Wrong!")

        print(f"Correct answer: {constellation.name}")

    def start(self):
        """Start the game"""
        print("✨ Welcome to the Constellation Guessing Game!")

        rounds = 3

        for i in range(rounds):
            print(f"\n--- Round {i+1} ---")
            self.play_round()

        print(f"\n🎯 Final score: {self.score}/{rounds}")


# =========================
# MAIN
# =========================
if __name__ == "__main__":
    game = GuessGame(constellation_data)
    game.start()
```
