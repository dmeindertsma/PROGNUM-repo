```python
from math import *
from scipy.constants import G
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
Solars={}

for M,r,n in zip(masses,radii,names):
    v_esc=sqrt((2*G*M)/r)
    print(f"{n} has the mass {M:.2e} and the escape velocity {v_esc}")
    Solars[n]=v_esc

print(Solars)

```

```python
import numpy as np
from math import *
xarray=[1,2,3,4]
xnparray=np.array(xarray)

def stdev(y):
    Sum=sum(y)
    N=len(y)
    mean=Sum/N
    
    sigma=sqrt((1/N)*sum((y-mean)**2))
    return sigma
standarddev=stdev(xnparray)
print(standarddev)

standarddev2=np.std(xnparray)
print(standarddev2)

def stdevoneline(z):
    sigma=sqrt((1/len(z))*sum((z-(sum(z)/len(z)))**2))
    return sigma
standarddev3=stdevoneline(xnparray)
print(standarddev3)
```
