```python
from math import *
from scipy.constants import G
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
Solars={}

for M,r,n in zip(masses,radii,names):
    v_esc=sqrt((2*G*M)/r)
    print(f"{n} has the mass {M:.2e} and the escape velocity {v_esc}")
    Solars[n]=v_esc

print(Solars)

```

```python
from math import *
d=float(input(f"d = "))
D=float(input(f"D = "))
def angular_size(d,D):
    delta_rad=2*atan(d/(2*D))
    delta=degrees(delta_rad)
    return delta
print(f"{angular_size(d,D)} degrees")
```

```python
from scipy.constants import pi
def cylinder(r,h):
    A=2*pi*r**2+2*pi*r*h
    V=pi*(r**2)*h
    return A, V
cylinder(1,2)
area, volume= cylinder(1,2)
print(area/volume)
```
