# Energy and frequency

With Planck's constant $h$, we write the relation between energy and frequency as:

$$E = hf$$

$$E = hc / \lambda$$

$$F=G \frac{m_1 m_2}{r^2}$$

$$\sin(2x)=2\sin(x)\cos(x)=\frac{2\tan(x)}{1+\tan^2(x)}$$

$$\sum_{k=0}^{\infty} \frac{1}{4k^2 - 1} = \frac{1}{2}$$
$$\int_{0}^{\pi} e^{ix} dx= 2i$$

```python
from math import *
alpha_d = 30                  # degrees
alpha_r = radians(alpha_d)    # Now in radians
print (sin(2*alpha_r))
print (2*sin(alpha_r)*cos(alpha_r))
print ((2*tan(alpha_r))/(1+(tan(alpha_r))**2))
```
