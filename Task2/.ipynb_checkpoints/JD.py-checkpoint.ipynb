```python
D=float(input("Day as a floating point number:"))
M=float(input("Month as a number:"))
Y=float(input("Year:"))
JD = 367*Y -7*(Y+(M+9)//12)//4 - 3*((Y+(M-9)//7)//100 + 1)//4 + (275*M)//9 + D + 1721029-0.5
print(f"The Julian Date for {D, M, Y} is: {JD}")
```
