```python
import numpy as np

# Defining R, P and S

R="R"
P="P"
S="S"

Start = input(f"Choose rock: R, paper: P or scissors: S: ")        # Players input
RPS = np.array(['R', 'P', 'S'])                                    
indx = np.random.randint(0, len(RPS), 1)                           # Random respons
print(RPS[indx])

# Stating the conditions of winning or losing:

if Start==R and RPS[indx]=='S':
    print("You win!")
elif Start==P and RPS[indx]=='R':
    print("You win!")
elif Start==S and RPS[indx]=='P':
    print("You win!")
elif Start==R and RPS[indx]=='R':
    print("Tie!")
elif Start==P and RPS[indx]=='P':
    print("Tie!")
elif Start==S and RPS[indx]=='S':
    print("Tie!")
else:
    print("You lose :(")
```

```python
import numpy as np
from matplotlib import pyplot as plt
lengths = np.random.normal(185, 8, 100000)
mean=np.mean(lengths); print(mean)
std=np.std(lengths); print(std)
plt.hist(lengths, bins=40, label=f"Lengths")
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean:.2f}")
plt.axvspan(mean - std, mean + std,color='green', alpha=0.2,label="±1 Std Dev")
plt.legend()
plt.show()

from scipy.stats import *
h=170
prob= norm.cdf(h, loc=mean, scale=std)
print(f"P(X ≤ {h}) = {prob:.4f}")

prob2= norm.cdf(mean + std, loc=mean, scale=std) - norm.cdf(mean - std, loc=mean, scale=std)
print(f"μ - σ ≤ X ≤ μ + σ = {prob2:.4f}")
```
