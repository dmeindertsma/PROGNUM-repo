```python
import numpy as np
from matplotlib import pyplot as plt
lengths = np.random.normal(185, 8, 100000)
mean=np.mean(lengths); print(mean)
std=np.std(lengths); print(std)
plt.hist(lengths, bins=40, label=f"Lengths")
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean:.2f}")
plt.axvspan(mean - std, mean + std,color='green', alpha=0.2,label="±1 Std Dev")
plt.legend()
plt.show()

from scipy.stats import *
h=170
prob= norm.cdf(h, loc=mean, scale=std)
print(f"P(X ≤ {h}) = {prob:.4f}")

prob2= norm.cdf(mean + std, loc=mean, scale=std) - norm.cdf(mean - std, loc=mean, scale=std)
print(f"μ - σ ≤ X ≤ μ + σ = {prob2:.4f}")
```

```python
import numpy as np

def funcfilter(func,seq):
    array=np.asarray(seq)
    filtered=[]
    for x in array:
        if x>0:
            filtered.append(x)
    filtered=np.asarray(filtered)      
    final=func(filtered)
    return final
    
sq = np.array([-100, -10, 0, 10, 100])
result = funcfilter(np.log10, sq)
print("Result log10:", result)

result2= funcfilter(np.log10, [100])
print("Result log10:", result2)
```

```python
from astropy.cosmology import Planck18
import numpy as np

H0 = Planck18.H0.value
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0

from scipy import integrate

def H(z):
    Hz=H0*(Om0*((1+z)**3)+Ode0)**0.5
    return Hz

def tL(z):
    integral=1/((1+z)*H(z))
    return integral

z=2

lookback_time= integrate.quad(tL,0,z)
print(lookback_time)                          # In (1/(km/s/Mpc))

# To get Gyr

Mpc_to_km=3.08567758128e19
s_to_Gyr=3.156e16
tLz_in_Gyr= np.array(lookback_time)*Mpc_to_km/s_to_Gyr

print(tLz_in_Gyr)
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import *

# Random arrays
X = np.random.normal(0, 1, 100000)
Y = np.random.normal(2, 1, 100000)

# PDF of random functions
x=np.linspace(-6,8,1000)
pdf_X= norm.pdf(x,0,1)
pdf_Y= norm.pdf(x,2,1)

plt.figure(figsize=(8,5))
plt.hist(X, bins=30, density=True, alpha=0.5, color='blue',edgecolor='black', label='X')
plt.hist(Y, bins=30, density=True, alpha=0.5, color='red', edgecolor='black',label='Y')
plt.plot(x,pdf_X,'k:', linewidth=5,label='X(PDF)')
plt.plot(x,pdf_Y,'k--', linewidth=5,label='Y(PDF)')
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.title("1D Histogram + PDF")
plt.annotate(r'$\mathcal{N}(0,1)$',
             xy=(-2.0, 0.15),     # waar de pijl naartoe wijst
             xytext=(-5.0, 0.15), # tekst links
             color='blue',
             verticalalignment='center',
             fontsize=18,
             arrowprops=dict(color='blue', lw=0, width=12,headwidth=30 ))
plt.annotate(r'$\mathcal{N}(2,1)$',
             xy=(4.0, 0.15),      # waar de pijl naartoe wijst
             xytext=(5.0, 0.15),  # tekst rechts
             color='red',
             verticalalignment='center',
             fontsize=18,
             arrowprops=dict(color='red', lw=0, width=12,headwidth=30))
plt.xlim(-6,8)

plt.xticks(range(-6,9,1))
plt.legend()
plt.grid(axis='y',linestyle='--', alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.normal(0,1,100000)
Y = np.random.normal(2,1,100000)

plt.figure(figsize=(8,6))

# histogram
H, xedges, yedges = np.histogram2d(X,Y,bins=30)             # Verdeelt in 30 bins

plt.hist2d(X,Y,bins=30,cmap='Blues')
plt.colorbar()

# normaliseren
H = H / H.sum()

# levels berekenen
Hflat = H.flatten()
Hsorted = np.sort(Hflat)[::-1]
cumsum = np.cumsum(Hsorted)

levels = []
for sigma in [0.68,0.95,0.997]:
    level = Hsorted[np.searchsorted(cumsum, sigma)]
    levels.append(level)
levels = np.sort(levels)

# bin centers
xcenters = (xedges[:-1] + xedges[1:]) / 2
ycenters = (yedges[:-1] + yedges[1:]) / 2


cs = plt.contour(xcenters, ycenters, H.T,
                 levels=levels,
                 colors='red',
                 linewidths=2)

labels = {
    levels[0]: "3σ",
    levels[1]: "2σ",
    levels[2]: "1σ"
}

plt.clabel(cs, fmt=labels, fontsize=16)

plt.xlabel("X")
plt.xticks(range(-5,5,2))
plt.ylabel("Y")
plt.title("2D Histogram + Contour")
plt.grid(True, linestyle='--', color='black')

plt.show()
```
